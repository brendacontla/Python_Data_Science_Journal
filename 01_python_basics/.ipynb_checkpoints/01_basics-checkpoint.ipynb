### **Exercise 1:** Sum or multiplication  

 Write a Python program that takes two integers and return their product if it is ≤ 1000. Otherwise, return their sum.

```python
# Step 1. Ask for the numnbers
n1 = int(input('Please enter an integer'))
n2 = int(input('Please enter an integer'))

# Step 2. Calculate their product
product = n1 * n2

# Step 3. Check if the product is lower or equal than 1000
if product <= 1000:
    r = product
else:
    r = n1 + n2

#Step 4. Print results
print("\n")
print(f'The result is {r}')
```

### **Exercise 2:** Sum of Current and Previous Number
Iterate through the first 10 numbers (0–9). For each number, print the sum of the current number and the previous number.

```python
# Step 1. Create a variable for indicating the previous number and summing
previous_number = 0 # The numbers must start in 0
total_sum = 0

# Step 2. Iteration and printing results
print('Printing current and previous number sum in a range(10)')
for i in range(0,10):
    # Calulate the sum
    total_sum += i
    # Get the previous number
    if i == 0:
        previous_number = i
    else:
        previous_number = i - 1
    # Results    
    print(f'Current number {i} Previous number {previous_number} Sum: {total_sum}')
```

### **Exercise 3:** Characters at Even Index
Ask the user to enter a string. Display the characters that are present at **even index positions** (0, 2, 4, …)

```python
# Step 1. Ask for the string to user
string_user = input('Enter a string')
print(f'Original string is {string_user}')

# Step 2. Printing characters at even positions
print('Printing only even index chars')

s_leng = len(string_user)

for i in range(0, s_leng):
    # Using modulo operator to know if the number is even
    check_even = i % 2 
    # Print only when it is an even number
    if check_even == 0:
        print(string_user[i])
    else:
        continue

# Note: I can use string[::2] or range(start, stop, step) for a faster solution!!
```

### **Exercise 9:** Check Palindrome Number
Check if a given number is a palindrome. A palindrome reads the same forwards and backwards. For example, 545 is a palindrome number.

```python
# Step 1. Get the number (I keep practicing input() so I'll ask the user)
number = input('Enter a number: ')

# Step 3. Reverse the number
reverse_number = number[::-1]

# Step 4. Print results
print("\n")
print(f'original number {number}')
if number == reverse_number:
    print(f'Yes. given number is palindrome number')
else:
    print(f'No. given number is not palindrome')
```

### **Exercise 10:** Merge Two Lists with Conditions
Given two lists of numbers:  
  - Create a **new list** containing **odd numbers** from the first list and **even numbers** from the second list  

```python
list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]

# Step 1. Get the even from list1 using list comprehensions (faster and cleaner)
even = [x for x in list1 if x % 2 == 0]
odds = [x for x in list2 if not x % 2 == 0]

# Step 2. Concatenate the lists 
complete_list = even + odds 
print(complete_list)
```

### **Exercise 11:** Reverse digits of a number
Write a program that prints the digits of a given integer in reverse order, separated by spaces.

```python
# Step 1. Ask for the given number
number = int(input('Enter a number:'))

# Step 2. Reverse the number 
div = number
while div != 0:
    residual = div % 10
    div = div // 10

    print(residual, end =" ")
```

### **Exercise 12:** Calculate income tax
Write a program to compute the income tax based on the rates given.

```python
# Step 1. Ask for the income
income = int(input('Enter the income to calculate the taxes'))

# Step 2. Calculate the tax
if income >= 20000 :
    tax = 10000 * 0.1 + (income - 20000) * 0.2
elif income >= 10000 and income < 20000:
    tax = (income - 10000) * 0.1
else:
    tax = 0

print(f'The tax is {tax}')
```

### **Exercise 13:** Multiplication table (1 to 10)
Write a program to generate and display the multiplication table for numbers 1 through 10.

```python
# Step 1. Define the index row
for i in range (1,11):
    print(i, end = '  ')
    # Step 2. Define the index column
    for j in range(2,11):
        m = i * j
        print(m, end = ' ')
    print('\n')
```

### **Exercise 14:** Downward half-pyramid of stars
Write a program to print the following star pattern:  
    * * * * *  
    * * * *  
    * * *  
    * *  
    *  

```python
pattern = '*'

for i in range(5,0,-1):
    for j in range(1,i+1):
        print(pattern, end = ' ')
    print('\n')
```
