### **Exercise 1:** Sum or multiplication  

 Write a Python program that takes two integers and return their product if it is ≤ 1000. Otherwise, return their sum.

```python
# Step 1. Ask for the numnbers
n1 = int(input('Please enter an integer'))
n2 = int(input('Please enter an integer'))

# Step 2. Calculate their product
product = n1 * n2

# Step 3. Check if the product is lower or equal than 1000
if product <= 1000:
    r = product
else:
    r = n1 + n2

#Step 4. Print results
print("\n")
print(f'The result is {r}')
```

### **Exercise 2:** Sum of Current and Previous Number
Iterate through the first 10 numbers (0–9). For each number, print the sum of the current number and the previous number.

```python
# Step 1. Create a variable for indicating the previous number and summing
previous_number = 0 # The numbers must start in 0
total_sum = 0

# Step 2. Iteration and printing results
print('Printing current and previous number sum in a range(10)')
for i in range(0,10):
    # Calulate the sum
    total_sum += i
    # Get the previous number
    if i == 0:
        previous_number = i
    else:
        previous_number = i - 1
    # Results    
    print(f'Current number {i} Previous number {previous_number} Sum: {total_sum}')
```

### **Exercise 3:** Characters at Even Index
Ask the user to enter a string. Display the characters that are present at **even index positions** (0, 2, 4, …)

```python
# Step 1. Ask for the string to user
string_user = input('Enter a string')
print(f'Original string is {string_user}')

# Step 2. Printing characters at even positions
print('Printing only even index chars')

s_leng = len(string_user)

for i in range(0, s_leng):
    # Using modulo operator to know if the number is even
    check_even = i % 2 
    # Print only when it is an even number
    if check_even == 0:
        print(string_user[i])
    else:
        continue
    
```

### **Exercise 4:** Remove First n Characters  
Remove the first `n` characters from the string and return the remaining string  

```python
# Step 1: Ask user the string and the value of n
string_user = input('Enter the string')
n = int(input('Enter the value of n'))

# Step 2: Printing the results
print('Original string:', string_user)
print(f"String with {n} characters removed: {string_user[n:]}")
```

### **Exercise 5:** Check First and Last Numbers in a List
Given a list of numbers, check whether the **first and last numbers** are the same. Return `True` if they are equal, otherwise return `False`.

```python
# Step 1: Define the list
numbers_list = [10, 20, 30, 40, 10]

# Step 2: Get the first and last numbers
first = numbers_list[0]
last = numbers_list[-1]

# Step 3: Compare them and return TRUE/FALSE to user
if first == last:
    print('True')
else:
    print('False')
```
