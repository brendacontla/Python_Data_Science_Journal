### **Exercise 2:** Sum of Current and Previous Number
Iterate through the first 10 numbers (0–9). For each number, print the sum of the current number and the previous number.

```python
# Step 1. Create a variable for indicating the previous number and summing
previous_number = 0 # The numbers must start in 0
total_sum = 0

# Step 2. Iteration and printing results
print('Printing current and previous number sum in a range(10)')
for i in range(0,10):
    # Calulate the sum
    total_sum += i
    # Get the previous number
    if i == 0:
        previous_number = i
    else:
        previous_number = i - 1
    # Results    
    print(f'Current number {i} Previous number {previous_number} Sum: {total_sum}')
```

### **Exercise 9:** Check Palindrome Number

### **Exercise 10:** Merge Two Lists with Conditions

### **Exercise 16:** Generate Fibonacci Series

Generate the Fibonacci series up to 15 terms.The Fibonacci series is a sequence where each number is the sum of the previous two numbers. The series starts with 0 and 1.

```python
# Step 1: Set the first two numbers
f_numbers = [0, 1] 

print('Fibonacci sequence:')
for i in range(0,13):
    # Step 2. Indicate the second number to sum to the first
    j = i + 1
    # Printing the first two numbers
    if i == 0:
        print(i, end = ' ')
        print(j, end = ' ')
    
    # Step 3. Summing the last and second last terms
    sum = f_numbers[i] + f_numbers[j]
    f_numbers.append(sum)
    print(sum, end = ' ')
```

### **Exercise 17:** Check Leap Year

Write a program to check whether a given year is a leap year. A leap year has 366 days.

Rules: A year is a leap year if it’s divisible by 4, unless it’s also divisible by 100 but not by 400. 

```python
# Step 1. Ask for the year
year = int(input('Enter the year to check:'))

# Step 2. Check if it's leap year
if not year % 4 == 0:
    print('Output False')
elif not year % 100 == 0:
    print('Output True')
elif year % 400 == 0:
    print('Output True')
else:
    print('Output False')
```

### **Exercise 18:** Print Alternate Prime Numbers

Print alternate prime numbers from 1 to 20. A prime number is divisible only by 1 and itself.

Prime numbers from 1 to 20:
2, 3, 5, 7, 11, 13, 17, 19

Alternate prime numbers:
2, 5, 11, 17


```python
# Step 1. Create a function to calculate if a number is a prime (Not very efficient if n is a large number!)
def is_prime(n):
    if n <= 1:
        return False
    else:
        for i in range(2, (n + 1)):
            div = n % i
            if div == 0 and i != n:
                return False
            else:
                return True

# Step 2. Identify the prime numbers in 1:20
p_numbers = []
for i in range (1, 21):
    if is_prime(i):
        p_numbers.append(i)

print(f'All prime numbers from 1 to 20: {p_numbers}')
print(f'Alternate prime numbes from 1 to 20: {p_numbers[::2]}')
```

### **Exercise 22:** Countdown Timer
Write a program that implements a simple countdown timer of five seconds using a `while` loop.
When the countdown reaches zero, display the message “Time’s up!”.

```python
n = 5

while n >= 0:
    # Checking the seconds
    if n > 0:
        print(f'Time remaining: {n} seconds')
        n -= 1
    else:
        print("Time's up!")
        break # Just to make sure it will stop
```

### **Exercise 11:** Percentage Display

Ask the user for a numerator and a denominator. Calculate the percentage and display it with two decimal places followed by a percent sign (e.g., 75.50%).

```python
# Step 1. Ask the user the numerator and denominator. A function is more convinient to use to check that the input is float.
def number_input():
    while True:
        number = input('Please enter a number:')

        try:
            number_f = float(number)
            break
        except ValueError:
            print('The input is not a number. Please enter a number.')

    return number_f

numerator = number_input()
denominator = number_input()

# Step 2. Calculate the percentage
p = numerator / denominator

# Step 3. Print the results
print(f'The percentage is: {p:.2%}')

# Note: It doesn't check with denominator 0!
```

- In Markdown, you need **two spaces at the end of a line** to force a line break  
- `print("\n")` adds a blank line in between output (improves readability in output)
- If I need elements at even indices, or every n-th element, I can use **slicing**: `string[::2]`. Alternatively, I can use **range** with a step: `range(start, stop, step)`
